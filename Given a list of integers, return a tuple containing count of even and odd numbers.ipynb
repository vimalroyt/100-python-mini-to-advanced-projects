```python
#Given a list of integers, return a tuple containing count of even and odd numbers.
```

```python
def even_odd_count(input_list):
    even_n=0
    odd_n=0

    for n in input_list:
        if n%2==0:
            even_n+=1
        else:
            odd_n+=1
    return even_n,odd_n

input_list=[1,2,3,5,4,7,85,9,8,84,8,9,78,46]
even_odd_count(input_list)
```
