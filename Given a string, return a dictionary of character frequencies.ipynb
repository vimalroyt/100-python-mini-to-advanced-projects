```python
#Given a string, return a dictionary of character frequencies (ignore spaces).

def character_frequency_count(input_string):
    char_freq={}

    for char in input_string:
        if char!=" ":
            if char in char_freq:
                char_freq[char]+=1  #If you're confused in this line, below i've explained it working.
            else:
                char_freq[char]=1
    return char_freq

input_string="I'm Vimal, i am good"
character_frequency_count(input_string)

```
