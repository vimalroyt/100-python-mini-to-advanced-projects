```python
#Write a function to flatten a nested list.
def flatten(lst):
    result=[]
    
    for i in lst:
        if isinstance(i,list):        # extend() method is used to add multiple items from an iterable (such as a list, tuple, string, or set) to the end of an existing list.
            result.extend(flatten(i)) # isinstance(object, classinfo)-The isinstance() function in Python is a built-in function used to check if an object is an instance of a specified class, type, or any of its subclasses. It returns True if the check is successful and False otherwise, making it a valuable tool for type checking and error handling in your code. 
        else:                         # object: The variable or object you want to check.
            result.append(i)          # classinfo: A class, type, or a tuple of classes and types to compare against the object's type. 
    return result

nested_list=[1,2,[3,4],5,6]
print(flatten(nested_list))

#Note
# flatten(): Creates a new array (copy). It is safer if you want to avoid changing the original.
# ravel(): Returns a view of the original data whenever possible. It is much faster and uses less memory, but changes to the result will change the original array. 
```
